```python
def get_list(path,loss=False):
    lis=[]
    with open(path,'r') as f:
        thing=f.readlines()
        
    for item in thing:
        if(loss==False):
            lis.append(eval(item).cpu().item())
        if(loss==True):
            lis.append(eval(item))
    return lis
```
